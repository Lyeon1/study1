문제 5) 방자의 영어 점수를 80점으로 수정하고, 평균 점수도 다시 계산해보세요.

import pandas as pd

data = {
    "국어": [80, 90, 70, 30],
    "영어": [90, 70, 60, 40],
    "수학": [90, 60, 80, 70],
}

columns = ["국어", "영어", "수학"]

index = ["춘향", "몽룡", "향단", "방자"]

df = pd.DataFrame(data, index=index, columns=columns)

```python
data = { "국어": [80, 90, 70, 30], "영어": [90, 70, 60, 40], "수학": [90, 60, 80, 70], }
columns = ["국어", "영어", "수학"]
index = ["춘향", "몽룡", "향단", "방자"]
df = pd.DataFrame(data, index=index, columns=columns)

df.loc[['방자'],['영어']]=80

과목=list(df.columns)
과목수=len(과목)
df['평균']=0

for i in 과목:
    df['평균']+=df['{}'.format(i)]

df['평균']/=과목수

df
```
