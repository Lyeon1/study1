```python
ebola=pd.read_csv('/content/country_timeseries.csv')
ebola
```

```python
ebola2=pd.melt(ebola,id_vars=['Date','Day'])
ebola2
```

```python
ebola_split=ebola2['variable'].str.split('_')
ebola_split
```

```python
state=ebola_split.str[0]
state
```

```python
country=ebola_split.str[1]
country
```

```python
ebola2['state']=state
ebola2['country']=country
del ebola2['variable']
ebola2
```

행에서 열로

```python
weather=pd.read_csv('/content/weather.csv')
weather
```

```python
weather2=pd.melt(weather,id_vars=['id','year','month','element',],
                 var_name='day',value_name='temp')
weather2
```

```python
weather3=weather2.pivot_table(index=['id','year','month','day'],
                     columns='element',
                     values='temp',
                     dropna=False)
weather3
```

```python
weather3=weather2.pivot_table(index=['id','year','month','day'],
                     columns='element',
                     values='temp',
                     dropna=True)
weather3
```

```python
weather4=weather3.reset_index()
weather4
```
